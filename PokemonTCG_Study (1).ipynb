```python
'''
This project is currently just exploratory. It's made to satisfy my own questions
about the Pokemon Trading Card Game in its entirety.
'''
# All imports
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display


# File being used
data = "pokemon.csv"
# Accessing file
df = pd.read_csv(data)
```

```python
# Wrangling data (removing, filling null, etc.)
def wrangle(filepath):

  #Setting index
  df = pd.read_csv(filepath, parse_dates = ["release_date"])
  # Sort the DataFrame by "release_date"
  df.sort_values(by="release_date", inplace=True)

  # Set the index to the "id" column
  df.set_index("release_date", inplace=True)

  # Cleaning, Filling, Dropping the specified columns
  columns_to_drop = ['id','evolvesFrom','evolvesTo','set','flavorText',
                       'subtypes','abilities','attacks','weaknesses','retreatCost',
                       'convertedRetreatCost','legalities','rules','regulationMark',
                       'ancientTrait','legalities','resistances','level']
  df.drop(columns=columns_to_drop, inplace=True)


  # Editing column names
  new_cols= {
    'supertype': 'cardtype',
    'nationalPokedexNumbers': 'nationalDexNum'
  }

  # Rename the columns
  df.rename(columns=new_cols, inplace=True)
  df = df[df['cardtype'] == 'Pok√©mon']
  df.drop(columns=['cardtype'], inplace=True)
  return df

# Removing all cardtype except pokemon
df = wrangle(data)
display(df.head(2))
```

```python
#Work in progress

def clean_types(value):
    return re.sub(r'[^a-zA-Z]', '', value)

# Creating a function to clean the ".0" from hp and the []
def clean_hp(value):
    if isinstance(value, list) and len(value) > 0:
        value = value[0]
        value = int(float(value))
    return value

# Applying the clean_types function
df['types'] = df['types'].apply(clean_types)

# Applying the clean_types function
df['nationalDexNum'] = df['nationalDexNum'].apply(clean_hp)

# Applying the clean_hp function
df['hp'] = df['hp'].apply(clean_hp)
```
