```python
'''
This will be a description of the file
'''
# All imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display


# File being used
data = "pokemon.csv"
# Accessing file
df = pd.read_csv(data)
```

```python
# Wrangling data (removing, filling null, etc.)
def wrangle(filepath):

  #Setting index
  df = pd.read_csv(filepath, parse_dates = ["release_date"])
  # Sort the DataFrame by "release_date"
  df.sort_values(by="release_date", inplace=True)

  # Set the index to the "id" column
  df.set_index("release_date", inplace=True)

  # Cleaning, Filling, Dropping the specified columns
  columns_to_delete = ['id','evolvesFrom','evolvesTo','set','flavorText',
                       'subtypes','abilities','attacks','weaknesses','retreatCost',
                       'convertedRetreatCost','legalities','rules']
  df.drop(columns=columns_to_delete, inplace=True)


  #Editing column names all at once
  new_cols= {
    'supertype': 'cardtype',
    'nationalPokedexNumbers': 'nationalDexNum'
  }

  # Rename the columns
  df.rename(columns=new_cols, inplace=True)
  df = df[df['cardtype'] == 'Pok√©mon']
  df.drop(columns=['cardtype'], inplace=True)
  return df

# Removing all cardtype except pokemon
df = wrangle(data)
display(df.head(2))
```
